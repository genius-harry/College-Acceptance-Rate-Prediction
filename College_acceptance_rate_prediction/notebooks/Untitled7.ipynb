```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
import matplotlib.pyplot as plt

# Load datasets
dataset_A = pd.read_csv('dataset_A.csv')
dataset_B = pd.read_csv('dataset_B.csv')

# Copy original datasets for non-standardized version
original_dataset_A = dataset_A.copy()
original_dataset_B = dataset_B.copy()

# Standardize numeric columns
scaler = MinMaxScaler()
numeric_columns = dataset_A.select_dtypes(include=[np.number]).columns

dataset_A[numeric_columns] = scaler.fit_transform(dataset_A[numeric_columns])
dataset_B[numeric_columns] = scaler.transform(dataset_B[numeric_columns])

# Columns with missing values
cols_with_na = [
    "SAT Math 75th percentile score",
    "ACT Composite 75th percentile score",
    "Endowment assets (year end) per FTE enrollment (FASB)"
]

def build_and_train_model(X_train, y_train, X_test, y_test, column_name):
    model = Sequential([
        Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
        Dropout(0.2),
        Dense(32, activation='relu'),
        Dropout(0.2),
        Dense(1)
    ])

    model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae'])

    history = model.fit(X_train, y_train, epochs=100, batch_size=32, validation_split=0.2, verbose=1)

    results = model.evaluate(X_test, y_test)
    print(f'Test loss for {column_name}: {results[0]}')
    print(f'Test MAE for {column_name}: {results[1]}')

    plt.figure(figsize=(12, 6))
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title(f'Model loss for {column_name}')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Validation'], loc='upper right')
    plt.show()

    y_pred = model.predict(X_test)

    plt.figure(figsize=(12, 6))
    plt.scatter(y_test, y_pred, alpha=0.3)
    plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
    plt.xlabel('Actual')
    plt.ylabel('Predicted')
    plt.title(f'Actual vs Predicted for {column_name}')
    plt.show()

    return model, y_pred

predictions = {}
for col in cols_with_na:
    print(f"Training model for {col}")

    X = dataset_A.drop(columns=cols_with_na + ['UniversityName'])
    y = dataset_A[col]
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)

    model, y_pred = build_and_train_model(X_train, y_train, X_test, y_test, col)
    predictions[col] = model.predict(dataset_B.drop(columns=cols_with_na + ['UniversityName']))

    model.summary()

# Fill missing values in non-standardized version
for col, pred in predictions.items():
    original_dataset_B[col].fillna(pd.Series(pred.flatten()), inplace=True)

# Combine standardized and non-standardized datasets
standardized_combined_dataset = pd.concat([dataset_A, dataset_B], ignore_index=True)
non_standardized_combined_dataset = pd.concat([original_dataset_A, original_dataset_B], ignore_index=True)

# Save datasets
standardized_combined_dataset.to_csv('standardized_combined_dataset.csv', index=False)
non_standardized_combined_dataset.to_csv('non_standardized_combined_dataset.csv', index=False)

# Check for missing values in combined datasets
missing_values_summary_std = standardized_combined_dataset.isna().sum()
missing_values_summary_non_std = non_standardized_combined_dataset.isna().sum()

print("Summary of missing values in standardized combined dataset:")
print(missing_values_summary_std)

print("Summary of missing values in non-standardized combined dataset:")
print(missing_values_summary_non_std)

remaining_na_columns_std = missing_values_summary_std[missing_values_summary_std > 0].index.tolist()
remaining_na_columns_non_std = missing_values_summary_non_std[missing_values_summary_non_std > 0].index.tolist()

if remaining_na_columns_std:
    print(f"Columns with missing values in standardized dataset: {remaining_na_columns_std}")
else:
    print("No missing values in the standardized combined dataset.")

if remaining_na_columns_non_std:
    print(f"Columns with missing values in non-standardized dataset: {remaining_na_columns_non_std}")
else:
    print("No missing values in the non-standardized combined dataset.")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from transformers import ViTFeatureExtractor, TFViTForImageClassification
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns

# Load your data (Assuming `selected_data` is already loaded)
# selected_data = pd.read_csv('path_to_your_data.csv')

# Drop the UniversityName column
# selected_data = selected_data.drop(columns=['UniversityName'])

# Normalize the data
scaler = MinMaxScaler()
numeric_columns = selected_data.select_dtypes(include=[np.number]).columns
selected_data[numeric_columns] = scaler.fit_transform(selected_data[numeric_columns])

# Ensure there are no NaN values
selected_data = selected_data.dropna()

# Convert the data into a matrix format suitable for CNNs
cnn_matrix = selected_data.to_numpy()

# Calculate the nearest perfect square greater than the number of columns
num_cols = cnn_matrix.shape[1]
next_perfect_square = int(np.ceil(np.sqrt(num_cols))**2)

# Pad the data with zeros to match the perfect square size
padding_size = next_perfect_square - num_cols
cnn_matrix_padded = np.hstack((cnn_matrix, np.zeros((cnn_matrix.shape[0], padding_size))))

# Reshape the data into a 2D format
reshape_size = int(np.sqrt(next_perfect_square))
cnn_matrix_reshaped = cnn_matrix_padded.reshape((cnn_matrix_padded.shape[0], reshape_size, reshape_size, 1))

# Ensure correct data types
cnn_matrix_reshaped = cnn_matrix_reshaped.astype(np.float32)
selected_data['AcceptanceRate'] = selected_data['AcceptanceRate'].astype(np.float32)

# Split the data into training and testing sets
train_data, test_data, y_train, y_test = train_test_split(cnn_matrix_reshaped, selected_data['AcceptanceRate'], test_size=0.2, random_state=123)

# Convert single-channel to three-channel (as ViT expects 3 channels)
train_data = np.repeat(train_data, 3, axis=-1)
test_data = np.repeat(test_data, 3, axis=-1)

# Ensure the images have the correct shape for ViT (224, 224, 3)
train_data = tf.image.resize(train_data, (224, 224))
test_data = tf.image.resize(test_data, (224, 224))

# Initialize the feature extractor and model
feature_extractor = ViTFeatureExtractor.from_pretrained('google/vit-base-patch16-224')
vit_model = TFViTForImageClassification.from_pretrained('google/vit-base-patch16-224')

# Preprocess the images
def preprocess_images(images):
    images = tf.image.convert_image_dtype(images, tf.float32)
    return feature_extractor(images, return_tensors="tf")["pixel_values"]

# Prepare the dataset
train_images = preprocess_images(train_data)
test_images = preprocess_images(test_data)

# Create TensorFlow datasets
train_dataset = tf.data.Dataset.from_tensor_slices((train_images, y_train)).batch(32)
test_dataset = tf.data.Dataset.from_tensor_slices((test_images, y_test)).batch(32)

# Build the custom model
inputs = tf.keras.Input(shape=(224, 224, 3), name="inputs")
vit_outputs = vit_model.vit(inputs)[0]  # Use vit_outputs[0] to get the correct output tensor
x = tf.keras.layers.GlobalAveragePooling1D()(vit_outputs)
x = tf.keras.layers.Dense(1024, activation='relu')(x)
x = tf.keras.layers.Dropout(0.5)(x)
x = tf.keras.layers.Dense(512, activation='relu')(x)
x = tf.keras.layers.Dropout(0.5)(x)
outputs = tf.keras.layers.Dense(1, activation='linear')(x)

model = tf.keras.Model(inputs=inputs, outputs=outputs)

# Compile the model
optimizer = tf.keras.optimizers.Adam(learning_rate=1e-4)
model.compile(optimizer=optimizer, loss=tf.keras.losses.MeanSquaredError(), metrics=[tf.keras.metrics.MeanAbsoluteError()])

# Train the model
history = model.fit(train_dataset, epochs=50, validation_data=test_dataset)

# Plot the training history
plt.plot(history.history['loss'], label='loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()

plt.plot(history.history['mean_absolute_error'], label='mean_absolute_error')
plt.plot(history.history['val_mean_absolute_error'], label='val_mean_absolute_error')
plt.xlabel('Epoch')
plt.ylabel('Mean Absolute Error')
plt.legend()
plt.grid(True)
plt.show()

# Evaluate the model on the test set
evaluation = model.evaluate(test_dataset)
print(f'Test loss: {evaluation[0]}')
print(f'Test MAE: {evaluation[1]}')

# Make predictions on the test set
predictions = model.predict(test_dataset)

# Plot the predicted vs actual acceptance rates
results = pd.DataFrame({'Predicted': predictions.flatten(), 'Actual': y_test})

# Ensure the plot's x and y axes are scaled correctly
min_val = min(results.min())
max_val = max(results.max())

sns.scatterplot(x='Actual', y='Predicted', data=results)
plt.plot([min_val, max_val], [min_val, max_val], color='red')
plt.title('Predicted vs Actual Acceptance Rates')
plt.xlabel('Actual Acceptance Rate')
plt.ylabel('Predicted Acceptance Rate')
plt.grid(True)
plt.show()
```
